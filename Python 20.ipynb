```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Array:", arr)
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))
```

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

C = A + B

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Addition:\n", C)
```
