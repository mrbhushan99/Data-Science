```python
# to convert the binary number into integer
num = int(input("enter 4 digit  binary number"))
p= 0 
decimal_no = 0
while num>0:
    digit = num%10
    num=num//10
    decimal_no += digit *(2**p)
    p+=1   
print(decimal_no)
```
