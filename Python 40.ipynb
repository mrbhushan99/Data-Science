```python
import pandas as pd

data = {
    "Name": ["Ram", "Shyam", "Aman"],
    "Marks": [80, 65, 90]
}

df = pd.DataFrame(data)

sorted_df = df.sort_values(by="Marks")

print(sorted_df)
```

```python
data = {
    "Marks": [70, 85, 90, 60]
}

df = pd.DataFrame(data)

print("Maximum:", df["Marks"].max())
print("Minimum:", df["Marks"].min())
```

```python
data = {
    "City": ["Pune", "Mumbai", "Pune", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)

print(df["City"].value_counts())
```

```python
data = {
    "Name": ["Ram", "Shyam"],
    "Age": [21, 22],
    "Marks": [80, 75]
}

df = pd.DataFrame(data)

df = df.drop("Age", axis=1)

print(df)
```

```python
data = {
    "Name": ["Ram", "Shyam", "Aman"],
    "Marks": [80, 75, 90]
}

df = pd.DataFrame(data)

print(df.iloc[1])
```
