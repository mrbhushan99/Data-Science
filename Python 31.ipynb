```python
import numpy as np

A = np.array([1,2,3])
B = np.array([4,5,6])

result = A * B

print("Result:", result)
```

```python
A = np.array([1,2,3])
B = np.array([4,5,6])

C = A + B

print("Addition:", C)
```

```python
A = np.array([10,20,30])
B = np.array([5,10,15])

C = A - B

print("Subtraction:", C)
```

```python
A = np.array([2,4,6])
B = np.array([3,5,7])

C = A * B

print("Multiplication:", C)
```

```python
A = np.array([10,20,30])
B = np.array([2,4,5])

C = A / B

print("Division:", C)
```

```python
arr = np.array([10,20,30,40,50])

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

```python
arr = np.array([5,2,8,1,9])

sorted_arr = np.sort(arr)

print("Sorted Array:", sorted_arr)
```

```python
arr = np.array([10,45,22,67,34])

index = np.argmax(arr)

print("Index of maximum element:", index)
```

```python
A = np.array([1,2,3])
B = np.array([4,5,6])

C = np.concatenate((A,B))

print("Concatenated Array:", C)
```

```python
arr = np.array([1,2,3,4,5])

rev = arr[::-1]

print("Reversed Array:", rev)
```

```python
arr = np.array([1,2,2,3,4,4,5])

unique = np.unique(arr)

print("Unique Elements:", unique)
```
