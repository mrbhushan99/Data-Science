```python
# Program to check if a number is even or odd

num = int(input("Enter a number: "))

if num % 2 == 0:
    print("The number is even")
else:
    print("The number is odd")
```

```python
# Program to check if a number is positive, negative, or zero

num = int(input("Enter a number: "))

if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero")
```

```python
# accessing elements in the list 

fruits = ["apple", "banana", "cherry"]

print(fruits[0])    # apple
print(fruits[1])    # banana
print(fruits[-1])   # cherry (last item)
```

```python
#modifying the list 
fruits = ["apple", "banana", "cherry"]
fruits[1] = "orange"
print(fruits)   # ['apple', 'orange', 'cherry']
```

```python
#List Operations
#1. Adding Elements
fruits = ["apple", "banana"]

# Append
fruits.append("cherry")  # ['apple', 'banana', 'cherry']

# Insert at position
fruits.insert(1, "orange")  # ['apple', 'orange', 'banana', 'cherry']
#####  MY PRACTICE ########## 

names=['bhushan','sudesh','sahil']
# append
names.append('omkar')
print(names) #['bhushan', 'sudesh', 'sahil', 'omkar']

# Insert at position
names.insert(1,'sahil patil')
print(names)
```

```python
# removing elements 
fruits = ["apple", "banana", "cherry", "orange"]

# Remove by value
fruits.remove("banana")

# Remove last item
fruits.pop()

# Delete by index
del fruits[0]

print(fruits)


### MY PRACTICE 

fruits1=['apple1','banana2','cheery3','orange4']
print(fruits1)

fruits1.remove('apple1')
print(fruits1)

# delete by index 
del fruits1[1]
print(fruits1)

# remove last item
fruits1.pop()
print(fruits1)
```

```python
#List slicing 
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])   # [20, 30, 40]
print(numbers[:3])    # [10, 20, 30]
print(numbers[-2:])   # [40, 50]
```

```python
#Looping Through a List
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

```python
# Length of a List
print(len(fruits))  # Number of items
```

```python
#sorting or reversing 
numbers = [5, 3, 8, 1]

numbers.sort()       # [1, 3, 5, 8]
numbers.reverse()    # [8, 5, 3, 1]
```
