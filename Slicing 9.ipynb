```python
s = "fullstackdevelopment"
```

```python
word = "stack"
rev = word[::-1]

result = s.replace(word, rev)

print(result)
```

```python
result = s[-8:-1][::-1]

print(result)
```

```python
first = s[:6][::-1]
middle = s[6:-6]
last = s[-6:][::-1]

result = first + middle + last

print(result)
```

```python
rev = s[::-1]
result = rev[::2]

print(result)
```

```python
mid = len(s)//2

result = s[mid:] + s[:mid]

print(result)
```
