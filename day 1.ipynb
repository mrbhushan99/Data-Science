```python

sentence = "Data Science is powerful" 
words =  sentence.split()

result = " "

for word in words:

    if len(word) % 2 == 0:

        result += word[::-2] + " "

    else:

        result += word[::2] + " "

print(result)
```

```python
word = "understanding"

output = " "

for i in range(1, len (word), 2):

    if word[i] not in ['a', 'e', 'i', 'o', 'u']:

        output += word[i]

print(output)
```

```python
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.datasets import make_classification
    from sklearn.model_selection import train_test_split
    from sklearn.metrics import accuracy_score
    from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier

    # Generate synthetic dataset
    X, y = make_classification(n_samples=1000, n_features=20, random_state=42)

    # Split into training and testing
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Create models
    model1 = BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth=1), n_estimators=50)
    model2 = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=1), n_estimators=50)

    # Train models
    model1.fit(X_train, y_train)
    model2.fit(X_train, y_train)

    # Predictions
    pred1 = model1.predict(X_test)
    pred2 = model2.predict(X_test)

    # Accuracy
    print("Bagging Accuracy:", accuracy_score(y_test, pred1))
    print("Boosting Accuracy:", accuracy_score(y_test, pred2))

```
