```python
from abc import ABC, abstractmethod

class Payment(ABC):
    
    @abstractmethod
    def pay(self, amount):
        pass

class CreditCard(Payment):
    
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")

class UPI(Payment):
    
    def pay(self, amount):
        print(f"Paid {amount} using UPI")

p1 = CreditCard()
p2 = UPI()

p1.pay(1000)
p2.pay(500)
```
