```python
# wap to print multiplication table for n 
n = int(input("Enter value of n"))
i=1
while i<=10:
    print(i*n)
    i+=1
```

```python
# to find the sum of n natural numbers 
n = int(input("Enter value of n"))
i=1
sum= 0
while i<=n:
    sum+=i
    i+=1
print(sum)
```

```python
# to find the product of n natural numbers 
n = int(input("Enter value of n"))
i=1
product= 1
while i<=n:
    product*=i
    i+=1
print(product)
```
