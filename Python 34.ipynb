```python
import numpy as np

class DataAnalyzer:

    def __init__(self, data):
        self.data = np.array(data)

    def statistics(self):
        print("Mean:", np.mean(self.data))
        print("Median:", np.median(self.data))
        print("Standard Deviation:", np.std(self.data))

    def normalize(self):
        return (self.data - np.min(self.data)) / (np.max(self.data) - np.min(self.data))

data = [10, 20, 30, 40, 50, 60]

analyzer = DataAnalyzer(data)

print("Data:", analyzer.data)
analyzer.statistics()

print("Normalized Data:", analyzer.normalize())
```
