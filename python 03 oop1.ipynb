```python
# self keyword - reference of your current object 
class car():
    wheels = 4
    color = "white"
    seats = 7
    
    def start(self):
        print("car is running...")
    
    def stop(self):
        print("car has stoped..")
    
car1 = car()  # creating object of car class
car2 = car()
car3 = car()  # you can create as many objects of car as you want 
```

```python
print(car1.wheels)
print(car1.color)
print(car3.seats)
car2.start()
car1.stop()
```
