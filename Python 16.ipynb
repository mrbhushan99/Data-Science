```python
from abc import ABC, abstractmethod

class Employee(ABC):
    
    @abstractmethod
    def calculate_salary(self):
        pass

class FullTimeEmployee(Employee):
    
    def calculate_salary(self):
        print("Full Time Salary: 50000")

class PartTimeEmployee(Employee):
    
    def calculate_salary(self):
        print("Part Time Salary: 20000")

e1 = FullTimeEmployee()
e2 = PartTimeEmployee()

e1.calculate_salary()
e2.calculate_salary()
```
