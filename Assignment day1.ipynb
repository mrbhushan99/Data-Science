```python
'''*You are driving a little too fast, and a police officer stops you. Write a function to
return one of 3 possible results: "No ticket", "Small ticket", or "Big Ticket".
If your speed is 60 or less, the result is "No Ticket". If speed is between 61 and 80 inclusive, 
the result is "Small Ticket". If speed is 81 or more, the result is "Big Ticket".
Unless it is your birthday (encoded as a boolean value in the parameters of the function) -- on your birthday,
your speed can be 5 higher in all cases. **'''
```

```python
def check_speed(speed, is_birthday):
   
    if is_birthday:
        speed = speed - 5
    if speed <= 60:
        return "No Ticket"
    elif speed <= 80:
        return "Small Ticket"
    else:
        return "Big Ticket"
```

```python
print(check_speed(81, True))    
```

```python
print(check_speed(65, True))   
```

```python
print(check_speed(85, False))   
```

```python
print(check_speed(75, False))  
```

```python
print(check_speed(60, False))   # No Ticket
```

```python
'''Given a string print number num by extracting all the
digits from the string as in string
Print negative number if the first character in 
string is '-'
eg.
-123abcd should return -123,abcd456-->456,fdhfh78dsd89--->7889,-56dssd78-->-5678'''
```

```python
def extract_number(s):
    result = ''  # To store the digits

    # Checking if the string starts with a negative sign
    if s[0] == '-':
        result = '-'  # Adding negative sign at the start

    for char in s:
        if char.isdigit():  
            result += char  
    
    return result
```

```python
print(extract_number("-56dssd78"))     
```

```python
print(extract_number("fdhfh78dsd89"))
```

```python
print(extract_number("abcd456"))
```

```python
print(extract_number("-123abcd"))     
     
```

```python
'''Q2
WAP to check whether a number is palindrome or not on following conditions.
1. take input number
2. add number and its reverse
3. check that number is palindrome or not,if not then sum and its reverse'''
```

```python
def is_palindrome(n):
    return str(n) == str(n)[::-1]  
def reverse_number(n):
    return int(str(n)[::-1])  

num = int(input("Enter a number: "))

while True:
    rev = reverse_number(num)
    num = num + rev  
    if is_palindrome(num):
        print("Palindrome found:", num)
        break
```
