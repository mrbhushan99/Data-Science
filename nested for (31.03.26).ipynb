```python
# first approch 

# 100 200 35 40 60 this is ip list 
# output -- [335,235,400,395,375]
#335 =100+200+35
#235 =200 +35
#400 = 
input = [100,200,35,40,60]
sum_ = sum(input)
output = []
for i in input:
    temp =sum_ - i
    output.append(temp)
print(output)
# complexity O(n)
    
```
