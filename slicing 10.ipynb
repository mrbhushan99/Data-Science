```python
s = "cybersecurityengineering"
```

```python
word = "security"
rev = word[::-1]

result = s.replace(word, rev)

print(result)
```

```python
result = s[5:][::-1]

print(result)
```

```python
result = s[0] + s[1:-1][::-1] + s[-1]

print(result)
```

```python
result = s[::2] + s[::-2]

print(result)
```

```python
new = s[3:-3]

if new == new[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```
