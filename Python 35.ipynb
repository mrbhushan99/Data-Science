```python
import numpy as np

class ArrayAnalyzer:
    
    def __init__(self, arr1, arr2):
        self.arr1 = np.array(arr1)
        self.arr2 = np.array(arr2)

    def sum_arrays(self):
        return self.arr1 + self.arr2

    def difference(self):
        return self.arr1 - self.arr2

    def compare_max(self):
        print("Max of Array 1:", np.max(self.arr1))
        print("Max of Array 2:", np.max(self.arr2))

    def correlation(self):
        return np.corrcoef(self.arr1, self.arr2)

# input data
data1 = [10, 20, 30, 40]
data2 = [15, 25, 35, 45]

analyzer = ArrayAnalyzer(data1, data2)

print("Array 1:", analyzer.arr1)
print("Array 2:", analyzer.arr2)

print("Sum:", analyzer.sum_arrays())
print("Difference:", analyzer.difference())

analyzer.compare_max()

print("Correlation Matrix:\n", analyzer.correlation())
```
