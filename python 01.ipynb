```python
def calci(option, A, B):
    match option:
        case 1:
            print("Addition:", A+B)
        case 2:
            print("Subtraction:", A-B)
        case 3:
            print("Multiplication:", A*B)
        case 4:
            print("Division:", A/B)
        case 5:
            print("Floor Division:", A//B)
        case 6:
            print("Modulus:", A%B)
        case _:
            print("Invalid option")

option = int(input("Enter option: "))
A = int(input("Enter First number: "))
B = int(input("Enter Second number: "))

calci(option, A, B)
```
