```python
class ATM:
    def __init__(self, pin):
        self.__pin = pin

    def change_pin(self, old_pin, new_pin):
        if old_pin == self.__pin:
            self.__pin = new_pin
            print("PIN changed successfully")
        else:
            print("Incorrect old PIN")

    def verify_pin(self, pin):
        if pin == self.__pin:
            print("Access Granted")
        else:
            print("Access Denied")


atm = ATM(1234)
atm.verify_pin(1234)
atm.change_pin(1234, 5678)
atm.verify_pin(5678)
```
