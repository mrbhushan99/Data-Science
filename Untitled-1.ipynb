```python
class Animal:
    def speak(self):
        print("Animal makes a sound")


class Dog(Animal):
    def bark(self):
        print("Dog barks")


d = Dog()
d.speak()
d.bark()
```

```python
class Shape:
    def area(self):
        print("Area calculation")


class Circle(Shape):
    def area(self):
        print("Area of Circle = 3.14 x r²")


class Square(Shape):
    def area(self):
        print("Area of Square = side²")


c = Circle()
s = Square()

c.area()
s.area()
```
