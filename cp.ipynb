```python
# printing unique values present in the list 

a= [1,54,'hi','python',54,'hi']
out = []
i = 0
while i<len(a):
    if a[i] not in out:
        out.append(a[i])
    i+=1
print(out)
```

```python
# printing integers present in the list and add them to it 

a= [1,54,'hi','python',54,'hi']
out = []
i = 0
while i<len(a):
    if type(a[i]) == int:
        out.append(a[i])
    i+=1
print(out)
```

```python
# printing integers present in the list and add them to it
# but do not store duplicate integers

a= [1,54,'hi','python',54,'hi']
out = []
i = 0
while i<len(a):
    if type(a[i]) == int and a[i] not in out:
        out.append(a[i])
    i+=1
print(out)
```

```python
# str is present in list print ir 2 times 
a= [1,54,'hi','python',54,'hi']
out = []
i = 0
while i<len(a):
    if type(a[i]) == str :
        out.append(a[i]*2)
    i+=1
print(out)
```

```python
# print captial letters present in str or not 
a = "My Name Is Bhushan.. I Am From Kolhapur"
out = ''
i = 0
while i< len(a):
    if "A" < a[i] <"Z":
        out+=a[i]
    i+=1
print(out)
```

```python
# print captial letters present in str or not 
# another way 
a = "My Name Is Bhushan.. I Am From Kolhapur"
out = ''
i = 0
while i< len(a):
    if ord("A") < ord(a[i]) <ord("Z"):
        out+=a[i]
    i+=1
print(out)
```
