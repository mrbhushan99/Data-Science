```python
import numpy as np

class StudentMarks:
    def __init__(self, marks):
        self.marks = np.array(marks)

    def average(self):
        return np.mean(self.marks)

    def highest(self):
        return np.max(self.marks)

    def lowest(self):
        return np.min(self.marks)

    def result(self):
        if np.mean(self.marks) >= 40:
            return "Pass"
        else:
            return "Fail"

marks = [65, 78, 82, 55, 90]

student = StudentMarks(marks)

print("Marks:", student.marks)
print("Average:", student.average())
print("Highest:", student.highest())
print("Lowest:", student.lowest())
print("Result:", student.result())
```
