```python
class Triangle:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    def area(self):
        print(f"area of triangle with length {self.length} and hieght {self.breadth} is {0.5*(self.length)*(self.breadth)}")
        
class square:
    def __init__(self,side):
        self.side=side
        
    def area(self):
        print(f"area of square with side {self.side} is {self.side*self.side}") 
        
obTriangle=Triangle
obSquare=square

obTriangle=Triangle(3,4)
obTriangle.area()
obSquare=square(5)
obSquare.area()
    
    
```

```python
def fun(a,b):
    print(a+b)
    
def fun(a,b,c):
    print(a+b+c)
    
fun(2,3,4)
fun(4,5)  # here the recent definition of the function
          # i.e. 3 arguments is considered
        # recent 2 parameter function is overloaded with 3 parameter function
```

```python
# Encapsulation
"""
access modifiers:
public : accessed anywhere  inside and outside the class
private(_): accessed both inside and outeside the class 
# private variable can be used inside and outside in python bacause of pythons flexiblity 
protected(__): accessed only inside the class 
""" 
```

```python
class Students:
    ABC = "bhushan"
    _ABC= "satyajit"
    __XYZ="Aniruddh" # private variable can be used inside and outside in python bacause of pythons flexiblity 
    
    def protectedVar(self):
        print(self.__XYZ)
obStudent=Students()
obStudent.ABC
# public variable can accessed anywhere 
```

```python
obStudent._ABC
# private variable can be used inside and outside in python bacause of pythons flexiblity
```

```python
obStudent.__XYZ  # protected variable can not be used outside the class 
```

```python
obStudent.protectedVar() #accessing protected variable 
# defined function in the class and accessed this function using object (obStudent)
```
